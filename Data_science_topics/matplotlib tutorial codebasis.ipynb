```python
import matplotlib.pyplot as plt
%matplotlib inline
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
plt.plot(x,y,color='r',marker='+',linestyle='',markersize=20,alpha=1)
plt.xlabel('city')
plt.ylabel('temp')
plt.title('whether')
plt.show()
```

```python
day = [1,2,3,4,5,6,7]
max_t = [35,52,51,48,49,50,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
plt.plot(day,max_t,label='max')
plt.plot(day,min_t,label='min')
plt.plot(day,avg_t,label='avg')
plt.xlabel('day')
plt.ylabel('temp')
plt.title('whether')
plt.legend(loc=4,shadow= True,fontsize='small')
plt.grid(which='both',color='b',linestyle=':',linewidth=2)
plt.show()
```

```python
company = ['go','back','in','out']
revanue = [60,80,72,91]
profit = [20,30,32,50]
```

```python
y = np.arange(len(company))
y
```

```python
plt.xticks(y,company)
plt.bar(y-0.2,revanue,width=0.4,label='rev')
plt.bar(y+0.2,profit,width=0.4,label='pro')
plt.legend()
```
