```python
# findall, search, split, sub, findliter
#all ready use to re in complex method find the meta charcter
# Re.txt file as described in the video!
# Meta Characters
# [] A set of characters
# \ Signals a special sequence (can also be used to escape special characters)
# . Any character (except newline character)
# ^ Starts with
# $ Ends with
# * Zero or more occurrences
# + One or more occurrences
# {} Exactly the specified number of occurrences
# | Either or
# () Capture and group

import re

mystr = '''Tata Sons
        Tata Sons Private Limited is the holding company ai of Tata Group and holds the aii bulk of shareholding in 
        the Tata group of companies including their aii land holdings across India, tea estates and steel plants.
        The biggest two of these trusts are the Sir Dorabji Tata Trust and Sir Ratan Tata Trust.
        phone: 965455655677
        email: tarahr@gmail.com '''

# pattern = re.compile(r'land') #span to matlab ke aa land ni rang batavse
pattern = re.compile(r'.')  #any charecter
pattern = re.compile(r'.Trust') #kei baju hatu ae batavi didhu
pattern = re.compile(r'^Tata')  #start thy ke nahi batavava
pattern = re.compile(r'com$')  # end mu 6e ae jova mate
pattern = re.compile(r'ai*')  # jo ama apde ek sathe a and a+i ketl pan hoy kya location per hoy te batavse
pattern = re.compile(r'a*i*')  #ek sathe jetla charcter find karva hoy aeni pa6al multification muki devanu
pattern = re.compile(r'Tata+')  #ekla koi ek name aenu ketla 6 ane kya location ma 6e ae find mate plus apvanu
pattern = re.compile(r'ai{2}') #jo apde koi a=0 and i=1 hot apde bhargav me jetle sudhu nu find karva hoy pan thy
pattern = re.compile(r'(Tata){1}') #jo eklu tata serch lkarvu hatu atle and tatatata bane jode aevu serch karvu hoy to breckes ma 2 lakhavna
pattern = re.compile(r'ai{1} | fax') #1 ke atle ai j maliyu and tata atle tata maliyu and koi na hoy to nahi batavae
matches = pattern.finditer(mystr) 
for match in matches:   
    print(match)
# print(r'\n')
```

```python
# Special Sequences
# \A Returns a match if the specified characters are at the beginning of the string
# \b Returns a match where the specified characters are at the beginning or at the end of a word r"ain\b"
# \B Returns a match where the specified characters are present, but NOT at the beginning (or at the end) of a word

# \d Returns a match where the string contains digits (numbers from 0-9)
# \D Returns a match where the string DOES NOT contain digits
# \s Returns a match where the string contains a white space character
# \S Returns a match where the string DOES NOT contain a white space character
# \w Returns a match where the string contains any word characters (characters from a to Z, digits from 0-9, and the underscore _ character)
# \W Returns a match where the string DOES NOT contain any word characters   v
# \Z Returns a match if the specified characters are at the end of the string
import re

mystr = '''Tata Sons
        Tata Sons Private Limited is the holding company ai of Tata Group and holds the aii bulk of shareholding in 
        the Tata group of companies including their aii land holdings across India, tea estates and steel plants.
        The biggest two of these trusts are the Sir Dorabji Tata Trust and Sir Ratan Tata Trust.
        phone: +915455655677
        number : 45673-9876
        email: tarahr@gmail.com '''

pattern = re.compile(r'\ATata') #line starting kya thi thy ae mate
pattern = re.compile(r'\bTata') # ane ama koi location ma start ketlivar thy ae batava mate
pattern = re.compile(r'Tata\b') # ane tata name charcter ni koi pan location ma last ending thati hoy aena mate
pattern = re.compile(r'77\b')  #koi pan ni ending batava mate
pattern = re.compile(r'\d{5}-\d{4}') #specify the number of leva hoy tare ano use thy
# pattern = re.compile(r'')

matches = pattern.finditer(mystr) 
for match in matches:   
    print(match)
```
